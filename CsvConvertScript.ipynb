Skript na konverciu textových suborov dat

```python
import pandas as pd

#convert funkcia pre trenovacie datasety
def convert(input_file, output_file):
    with open(input_file, 'r') as f:
        lines = f.readlines()

    #pripravime si premenne
    data = []
    prompt = ''
    user = ''
    countries = ''
    days = ''
    client = ''
    session = ''
    form = ''
    time = ''
    sym = 0

    for line in lines:
        line = line.strip()
        #oddelime si metadata od dat
        if line.startswith('# prompt'):
            prompt = line.split('# prompt:')[1]
        elif line.startswith('# user'):
            line = line.split()
            user = line[1].split('user:')[1]
            countries = line[2].split('countries:')[1]
            days = line[3].split('days:')[1]
            client = line[4].split('client:')[1]
            session = line[5].split('session:')[1]
            form = line[6].split('format:')[1]
            time = line[7].split('time:')[1]
            sym = sym + 1
        else:
            items = line.split()
            #odstranime prompt pri cviceniach na pocuvanie
            if form == 'listen':
                prompt = ' '
            #zapiseme si data + metadata do stlpcov
            data.append(items + [prompt, user, countries, days, client, session, form, time] + [sym])

    columns = ['ID', 'token', 'pos', 'gramCat', 'morf', 'morf2', 'accuracy',  'prompt',
               'user', 'countries', 'days', 'client', 'session', 'format', 'time', 'sym']

    df = pd.DataFrame(data, columns=columns)
    df.to_csv(output_file + ".csv", index=False)
```

Použitie funkcie convert.

```python
convert('data/en_es_train.txt', 'data/en_es_train')
```

```python
convert('data/es_en_train.txt', 'data/es_en_train')
```

```python
convert('data/fr_en_train.txt', 'data/fr_en_train')
```
