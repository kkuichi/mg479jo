```python
#stlpce s gram kategoriami
check_cols = ['gramCat1', 'gramCat2', 'gramCat3', 'gramCat4', 'gramCat5', 'gramCat6']

#ideme po riadkoch
for index, line in es.iterrows():
    #if index == 0:
    #    continue
    #a teraz pozerame na stlpce
    for column in check_cols:

        #pridavne mena
        if es.loc[index, 'pos'] == 'ADJ':
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Gender='):
                es.loc[index, 'gender'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Number='):
                es.loc[index, 'number'] = es.loc[index, column].split('=')[1]

            #kategorie ktore nema
            es.loc[index, 'case'] = None
            es.loc[index, 'person'] = None
            es.loc[index, 'pronType'] = None
            es.loc[index, 'tense'] = None

        #aux
        if es.loc[index, 'pos'] == 'AUX':
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Number='):
                es.loc[index, 'number'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Person='):
                es.loc[index, 'person'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Tense='):
                es.loc[index, 'tense'] = es.loc[index, column].split('=')[1]
            #kategorie ktore nema
            es.loc[index, 'case'] = None
            es.loc[index, 'gender'] = None
            es.loc[index, 'pronType'] = None

        #det
        if es.loc[index, 'pos'] == 'DET':
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Gender='):
                es.loc[index, 'gender'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Number='):
                es.loc[index, 'number'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('PronType='):
                es.loc[index, 'pronType'] = es.loc[index, column].split('=')[1]

            #kategorie ktore nema
            es.loc[index, 'case'] = None
            es.loc[index, 'person'] = None
            es.loc[index, 'tense'] = None

        #podstatne mena
        if es.loc[index, 'pos'] == 'NOUN':
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Gender='):
                es.loc[index, 'gender'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Number='):
                es.loc[index, 'number'] = es.loc[index, column].split('=')[1]

            #kategorie ktore nema
            es.loc[index, 'case'] = None
            es.loc[index, 'person'] = None
            es.loc[index, 'pronType'] = None
            es.loc[index, 'tense'] = None

        if es.loc[index, 'pos'] == 'PRON':
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Case='):
                es.loc[index, 'case'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Gender='):
                es.loc[index, 'gender'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Number='):
                es.loc[index, 'number'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Person='):
                es.loc[index, 'person'] = es.loc[index, column].split('=')[1]

            #kategorie ktore nema
            es.loc[index, 'pronType'] = None
            es.loc[index, 'tense'] = None

        #slovesa
        if es.loc[index, 'pos'] == 'VERB':
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Number='):
                es.loc[index, 'number'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Person='):
                es.loc[index, 'person'] = es.loc[index, column].split('=')[1]
            if isinstance(es.loc[index, column], str) and es.loc[index, column].startswith('Tense='):
                es.loc[index, 'tense'] = es.loc[index, column].split('=')[1]
            #kategorie ktore nema
            es.loc[index, 'case'] = None
            es.loc[index, 'gender'] = None
            es.loc[index, 'pronType'] = None

```

Funkcia: jedno cvičenie == jeden riadok

```python
df = pd.read_csv('en_es_t.csv')
```

```python
df.head(10)
```

```python
group = df.groupby('sym')
new_df = (df
 .assign(col=group.cumcount().add(1).astype(str)
        )
 .pivot(index=['sym', 'prompt', 'user', 'countries', 'days', 'session', 'format', 'time'], columns='col')
 .pipe(lambda d: d.set_axis(d.columns.map('_'.join), axis=1))
 .reset_index()
)
```

```python
new_df.head(10)
```

```python
new_df.to_csv('transformed_en_es_t.csv', sep=',', index=False, encoding='utf-8')
```
